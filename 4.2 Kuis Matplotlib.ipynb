```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

## Soal 1
Dari data x dan y buatlah sebuah figure dengan ketentuan :
1. Judul figure adalah Data Penjualan PT. Sejahtera
2. X label adalah Pembelian
3. Y label adalah penjualan

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('Pembelian')
ax.set_ylabel('Penjualan')
ax.set_title('Data Penjualan PT.Sejahtera')
```

## Soal 3

Buatlah sebuah figure dengan ketentuan :
1. Menggunakan sub plot 1 baris 2 kolom

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

2. Sekarang isi data plot pertama dengan data x dan y dan buat sesuai dengan tampilan di bawah
3. Sekarang isi data plot pertama dengan data x dan z dan buat sesuai dengan tampilan di bawah

```python
axes[0].plot(x,y,color="blue", lw=3, ls='-',marker='$...$')
axes[1].plot(x,z,color="blue",lw=3,ls='-',marker='*')
fig
```

Buat tampilannya menjadi seperti ini

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))

axes[0].plot(x,y,color="blue", lw=3,ls='-',marker='$...$')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,z,color="blue", lw=3, ls='-',marker='*')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```
